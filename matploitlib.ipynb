```python
import numpy as np

x = np.linspace(0,5,11)
```

```python
x
```

```python
y = x**2
```

```python
import matplotlib.pyplot as plt
```

```python
plt.plot(x, y)
```

```python
fig = plt.figure()
axes = fig.add_axes([0.1, 0.1 , 0.8 , 0.8])
axes.plot(x,y , "red")

axes.set_xlabel("eixo x")

axes.set_ylabel("eixo y")

axes2 = fig.add_axes([0.3, 0.6 , 0.3 , 0.5])
axes2.plot(x,y , "red")

axes2.set_xlabel("eixo x")

axes2.set_ylabel("eixo y")
```

```python
fig, ax = plt.subplots()

ax.plot(x, y, "r--")

ax.set_title("title")
```

```python
fig, ax = plt.subplots(nrows=1, ncols=2)

for axis in ax:
  axis.plot(x, y, "r")

```

```python
fig, ax = plt.subplots(nrows=1, ncols=2)
ax[0].plot(x, y, "red")
ax[1].plot(y, x, "green")
```
