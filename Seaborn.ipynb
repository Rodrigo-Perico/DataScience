```python
import seaborn as sns
```

```python
tips = sns.load_dataset('tips')
```

```python
sns.distplot(tips['total_bill'])
```

```python
sns.jointplot(x='total_bill', y='tip', data= tips, kind= 'reg')
```

```python
sns.jointplot(x='total_bill', y='tip', data= tips, kind= 'hex')
```

```python
sns.pairplot(tips)
```

```python
sns.pairplot( data= tips, hue = 'sex')
```

```python
sns.kdeplot(tips['total_bill'])
```

```python
import numpy as np
sns.barplot(x='sex', y= 'total_bill', data=tips, estimator=np.std)
```

```python
sns.countplot(x= 'sex', data = tips)
```

```python
sns.boxplot(x = 'day', y= 'total_bill', data=tips)
```

```python
sns.violinplot(x = 'day', y= 'total_bill', data=tips, hue='sex', split=True)
```

```python
sns.lmplot(x = 'total_bill', y= 'tip', data=tips, hue='sex', markers= ['x', 'v'] )
```

```python
flights = sns.load_dataset('flights')
```

```python
flights
```

```python
df = flights.pivot_table(index = 'year', columns = 'month', values= 'passengers')
```

```python
import matplotlib.pyplot as plt
```

```python
fig, ax = plt.subplots(figsize=(12,7))

sns.heatmap(data = df, ax=ax, annot=True,fmt= '.0f')
```

```python
sns.set_style('white')

sns.countplot(x = 'sex', data=tips)
```

```python
sns.set_style('ticks')

sns.countplot(x = 'sex', data=tips, palette = 'deep')
```

```python
fig, ax = plt.subplots(figsize=(12,7))

sns.set_style('ticks')

sns.countplot(x='sex', data=tips, palette='deep', ax=ax)

ax.set_xlabel('eixo x')
ax.set_ylabel('eixo y')
ax.set_title('Ttulo')
```
