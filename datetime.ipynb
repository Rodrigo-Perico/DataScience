```python
import pandas as pd
```

```python
ndays = 100
```

```python
datas = pd.date_range(start = '1/1/2021', periods= ndays)
```

```python
datas
```

```python
df = pd.DataFrame(range(1, ndays+1), columns= ['number'], index = datas)
```

```python
df
```

```python
df.index.names = ['Data']
```

```python
df.index[40].hour
```

```python
df[df.index.day == 20]
```

```python
df["Month"] = df.index.month
```

```python
df["Year"] = df.index.year
```

```python
df["Day"] = df.index.day
```
