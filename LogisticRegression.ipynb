```python
from sklearn import datasets
import numpy as np
import matplotlib as plt

from sklearn.linear_model import LogisticRegression
```

```python
iris = datasets.load_iris()
```

```python
iris.keys()
```

```python
iris["data" ]
```

```python
iris["data"].shape
```

```python
x = iris["data"][:, 3:]
y = (iris["target"] == 2).astype(int)  # Substitua np.int por int
```

```python
logist = LogisticRegression()

logist.fit(x, y)
```

```python
fig, ax = plt.subplots(figsize= (20,6))

x_new = np.linspace(8, 3, 100).reshape(-1, 1)
y_proba = logist.predict_proba(x_new)
```
